**Сгенерировать 1000 случайных чисел с любым известным распределением 
и построить их гистограмму, математическое ожидание, дисперсию.**


```python
import scipy.stats as sts
```

#### Нормальное распределение
#### -----------------------

    * r = sts.norm.rvs(loc=a,scale=sigma, size=n_points) *

**отрисовывает распределение случайных точек по равномерному нормальному закону.
        a — параметр, характеризующий среднее значение
        sigma — параметр, характеризующий разброс параметров**




```python
numbers = sts.norm.rvs(loc = 0, scale = 1, size = 1000)
```

##### Generated 1000 random numbers based on Normal Distribution(Gauss) 

```python
print(numbers)
```

### Гистограмма

```python
plt.hist(numbers, density=True, histtype='stepfilled',alpha=0.8)
plt.title('Normal')
```

### Математическое ожидание


```python
print(f'Mean is {np.mean(numbers)}')
```

### Дисперсия


```python
print(f"Dispersion is {np.var(numbers)}")
```

```python
plt.hist(numbers , bins=100)
```

**Сгенерировать случайную точку, равномерно распределенную в квадрате 
со стороной а.**

```python
import random
```

### Генерируем случайную сторону

```python
r_number = random.randint(5, 10)
plt.xlim(r_number)
plt.ylim(r_number) 
```

#### Tочку в квадрате.


```python
dot = np.random.uniform(low = 1,high = r_number + 1, size= (1,2))

plt.plot(dot[:,0],dot[:,1],"ok")
```
